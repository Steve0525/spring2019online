# Looking at building inventory dataset
* reading, filtering, general mutation of data

```python
f = open("/Users/jillnaiman1/Downloads/building_inventory.csv")
```

```python
f.seek(0) # start at the top
for record in csv.reader(f):
    print(record)
```

```python
# lets make this a fancy dataset!
f.seek(0)
reader = csv.reader(f)
header = next(reader)
header
```

```python
data = {} # empty dict
for col in header:
    data[col] = []
data
```

```python
# we are going to use zip
a = ["hi", "there", "folks"]
b = [9, 4, 5]
for word, num in zip(a, b):
    print(word, num)
```

```python
# lets use zip to fill out dict
for row in reader:
    for col, val in zip(header, row):
        data[col].append(val)
data
```

```python
len(data['Zip code'])
```

```python
# lets say I forget what is in my dataset
data.keys()
```

```python
for i, location in enumerate(data['Address']):
    if '905 S Goodwin Ave' in location:
        print(data['Square Footage'][i])
```

# Now for something completely different: Manipulating Images!

```python
# pip3 install pillow
# conda install -c anaconda pillow
# anaconda GUI
import PIL.Image as Image

# now open an image
im = Image.open("/Users/jillnaiman1/Downloads/stitch_reworked.png","r")
```

```python
im.show()
```

```python
# lets work with the image data
import numpy as np
# might have to install with pip: pip3 install numpy
data = np.array(im)
data # each entry is a pixel of [r, g, b, alpha], alpha = transparency
```

```python
np.unique(data[:,:,0]) # unique values of R
```

```python
np.unique(data[:,:,1]) # unique values of G
```

```python
np.unique(data[:,:,2]) # unique values of B
```

```python
# practice with figures
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(5,5))
ax.imshow(im)
```

```python
#im.show()

# recall, this is a visualization!  
#  we can use this data/viz to quantify how good or bad stich is
# do this by volumetric pixel filling
ngood = (data[:,:,0] == 255).sum() # white, "good" parts
nbad = (data[:,:,0] == 126).sum() # red or "bad values"
total = ngood+nbad
```

```python
total, ngood, nbad
```

```python
badness = nbad/total
goodness = ngood/total
print(badness, goodness)
```

```python
# lets compare this to our "apparent" goodness
goodness_apparent = (50 - 150)/(50-450) # (top of image - badness level)/(top - full image)
goodness_apparent
```

```python
#plt.pie([badness, goodness])
fig,ax = plt.subplots(figsize=(5,5))
ax.pie([badness,goodness])
```

# Making diagrams in Python

```python
edgecolor = "#000000"
facecolor = "#1f77b4"

totalFOV = matplotlib.patches.Wedge([0.0, 0.0], 1.0, 90 - (210/2.0), 90 + (210/2.0), # span of the wedge
                                    lw=2.0, 
                                    facecolor=facecolor, edgecolor=edgecolor)
fig,ax = plt.subplots(figsize=(10,7), dpi=300)
ax.add_patch(totalFOV)
ax.set_xlim(-1.25,1.25)
ax.set_ylim(-0.5, 1.25)
plt.show()
```
