```python
# we are going to play with ipywidgets!
import ipywidgets
```

```python
# lets use widget!
# docs! https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html
@ipywidgets.interact(style=plt.style.available)
def make_plot(style):
    with plt.style.context(style):
        plt.plot(stats["max"],marker=".", linewidth=1.0, label="Max")
        plt.plot(stats["min"], marker=".", linewidth=1.0, label="Min")
        plt.fill_between(stats.index, stats["min"], stats["max"], color="#aaaaaa")
        plt.ylabel("Square Footage")
        plt.xlabel("Year")
        plt.yscale("log")
        plt.legend() 
```

```python
# lets say we want to fill the plot window instead of 
#  a datarange
import matplotlib.transforms as mpt
with plt.style.context("ggplot"):
    plt.plot(stats["max"],marker=".", linewidth=1.0, label="Max")
    plt.plot(stats["min"], marker=".", linewidth=1.0, label="Min")
    plt.fill_between(stats.index, stats["min"], stats["max"], color="#aaaaaa")
    plt.ylabel("Square Footage")
    plt.xlabel("Year")
    plt.yscale("log")
    plt.legend()
    
#fig, ax = plt.subplots(figsize=(5,5))
ax = plt.gca()

new_transform = mpt.blended_transform_factory(ax.transData, ax.transAxes)

# now lets draw our line in window units instead of data
plt.plot([1868,1868], [0.0, 1.0], color="black", linewidth=2, transform = new_transform)
```

```python
# now onto interactive binning of data!

# lets look at the mean square footage per year
mean_sq = stats["mean"]
mean_sq
```

```python
# lets do some interpolation... never mind!  We are doing some averaging over windows
@ipywidgets.interact(window=(1, 10, 1))
def make_plot(window):
    mean_sq_avg = mean_sq.rolling(window=window).mean() # take mean of values in each window
    plt.plot(mean_sq, marker='.', linewidth=0.5, alpha=0.5)
    plt.plot(mean_sq_avg, marker ='.', linewidth=1.5)
    plt.yscale("log")
```
